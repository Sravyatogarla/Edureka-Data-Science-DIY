**Q1. Problem Statement: Combining DataFrames -I**

Write a Python program with two DataFrames, with columns “name” and “age”. Now,  combine  both  the  DataFrames  vertically  using  the concat()function, without the duplicate rows, and print the final DataFrame

```python
#problem statement 1
import pandas as pd
#1 creating the dataframes
df1= pd.DataFrame({'name': ['Shubam','Adin','Rahul'],'age' :['23','28','30']})
display(df1)
df2 = pd.DataFrame({'name': ['Amrit','Shubam','Ali'],'age':['20','23','28']})
display(df2)
```

```python
#2 combining the two dataframes without duplicates
df3 = pd.concat([df1,df2]).drop_duplicates()
df3 = df3.reset_index(drop=True)
display(df3)
```

**Q2. Problem Statement: Combining DataFrames -II**

Write  a  Python  programto  create  two  DataFrames.  First  DataFrame  with columns “employee”and “department,”Second  DataFrame  with  columns “employee”and “joining_date,” and Perform the following tasks:1.Merge the DataFrames based on the “employee”column2.Display the final DataFrame with employees records who are working in the “Engineering” department

```python

#creating the two data frames
df1 = pd.DataFrame({'employee':['Lisa','Bob','Jack','Sue','John','Daisy'], 'joining_data' :['2004','2008','2012','2022','2021','2020']})
display(df1)
df2 = pd.DataFrame({'employee':['Bob','Jack','Lisa','Sue','John','Daisy'], 'department':['Accounting','Engineering','Engineering','HR','Engineering','Engineering']})
display(df2)
```

```python
#1 merge the Dataframes
df1 = pd.DataFrame({'employee':['Lisa','Bob','Jack','Sue','John','Daisy'], 'joining_data' :['2004','2008','2012','2022','2021','2020']})
df2 = pd.DataFrame({'employee':['Bob','Jack','Lisa','Sue','John','Daisy'], 'department':['Accounting','Engineering','Engineering','HR','Engineering','Engineering']})

import pandas as pd
df3 = pd.merge(df1,df2)
display(df3)
```

```python
df4 = df3.loc[df3['department'] =='Engineering']
display(df4)
```
