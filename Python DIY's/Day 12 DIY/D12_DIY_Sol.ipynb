**Q2. Problem Statement: Data Filtering using Pandas**

Use the above data set and do the below tasks.1.Remove the data of players who didn't score a single century2.Replace the player name "V Kohli " with "Virat Kohli “3.How many matches they weren't getting a chance for batting to fetch that and stored in a new column and named it as “not batting”4.Fetch  data  where  they  scored  neither  100  nor  50  in  matches  and store itin the new column as renamed as “low score innings”

```python

#read players based on the dataframe
df = pd.read_csv("player.csv")
#1 remove the data of players who didnt even score one centuary
df.drop(df[df['100s'] == 0] .index, inplace = True)
df
```

```python
#2 replace the name V kohli with Virat Kohli
df["name"] = df["name"].replace(to_replace="V Kohli", value="Virat Kohli")
df
```

```python
#3 create a column as "not batting" for how the matches they are not getting bating 
df["not batting"] = df["matches"] - df["innings"]
df
```

```python
#4 fetch the data where they neither scored 100s nor 50s store it in a new column as "low score innings"
df["low score innings"] = df["innings"] - df["100s"] - df["50s"]
df
```
