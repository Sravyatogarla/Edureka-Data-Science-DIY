Load the “cereal.csv”data into a DataFrame and perform the following tasks:

1.Explore the DataFrame using info()and describe()functions

2.Replace the manufacturer names with the dictionary of names given here –{'A':'American Home Food Products', 'G':'General Mills','K':'Kelloggs','N':'Nabisco', 'P':'Post', 'Q':'Quaker Oats', 'R':'Ralston Purina'}

3.Find out the top five cereal manufacturers based on customer ratings

4.With the help of a barplot, compare the manufacturers based on customer ratings

5.With  the  help  of  a pie  chart,  find  out  which  of  the  manufacturers  are  the largest producers of cereals

```python
# importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
# importing the csv file
df= pd.read_csv("cereal.csv")
```

```python
df
```

```python
#exploring the df using info() and describe() function
df.info()
df.describe()
```

```python
# replace manufacturers names usng the dictionary
manufacturer_dict = {"A" : "American Home Food products", "G" : "General Mills", "K": "Kelloggs","N": "Nabisco",
                     "P":"post", "Q":"Quaker Oats","R": "Ralston Purina"}
manufracture = df["mfr"].replace(manufacturer_dict)
print(manufracture)
```

```python
df
```

```python
# top 5 cereal manfacturers by customer rating
top_5_manufacturers = df.groupby("mfr")["rating"].mean().sort_values(ascending=False).head(5)
print(top_5_manufacturers)
```

```python
#Bar plot comparing manufacturers by rating
plt.figure(figsize=(10,6))
plt.bar(top_5_manufacturers.index, top_5_manufacturers.values)
plt.xlabel("Manufacturer")
plt.ylabel("Average Rating")
plt.title("Top 5 Cereal Manufacturers by Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
#pie chart of manufacturer market share
manufacturer_share = df["mfr"].value_counts().sort_values(ascending= False)
plt.figure(figsize= (8,8))
plt.pie(manufacturer_share, labels=manufacturer_share.index,autopct="%1.1f%%", startangle=140)
plt.title("Manufacturer market Share")
plt.show()
```
