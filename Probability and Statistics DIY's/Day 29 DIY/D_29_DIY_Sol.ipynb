**Q1. Problem Statement: Hypothesis Testing**

Write  a  Python  program  that  reads  the  Life  Expectancy  Data.csv  (provided  on LMS)  file  into  a  DataFrame,  then  drops  null  values  and performshypothesis testing for the given scenario.

H0:u= average life expectancy of Afghanistan is the same as the averageLife expectancy of the world (Null hypothesis)

H1:u= average life expectancy of Afghanistan is not the same as the average Life expectancy of the world (Alternate hypothesis)

1.Load the dataset into a data frame

2.Drop all missing values 

3.Take   the   mean   of   the   whole   target   variable   and   extract   data   from ‘Afghanistan’

4.consider  a  significance  level  of  0.05,  perform  a  T-tail  test  using  the  stats library and calculate P-value and conclude hypothesis testing.


```python
#importing all the neccessary libraries
import pandas as pd
from scipy.stats import ttest_1samp
```

```python
#1 Load the dataset
df = pd.read_csv("Life Expectancy Data.csv")
df
```

```python
#2 Drop missing values
df.dropna(inplace=True)
```

```python
#3 Extract life expectancy data for afghanistan
afghanistan_life_expectancy = df[df["Country"] == "Afghanisthan"] ["Life expectancy "]
afghanistan_life_expectancy
```

```python
# Calculate the mean life expectancy globally
global_life_expectancy_mean = df['Life expectancy '].mean()
global_life_expectancy_mean
```

```python
#4 Perform one-sample t-test
t_statistic, p_value = ttest_1samp(afghanistan_life_expectancy, global_life_expectancy_mean)
```

```python
# Set significance level
alpha = 0.05
```

```python
# Conclusion
if p_value < alpha:
    print("We reject the null hypothesis")
    print("The average life expectancy of Afghanistan is not the same as the average life expectancy of the world.")
else:
    print("We fail to reject the null hypothesis")
    print("The average life expectancy of Afghanistan is the same as the average life expectancy of the world.")
```
