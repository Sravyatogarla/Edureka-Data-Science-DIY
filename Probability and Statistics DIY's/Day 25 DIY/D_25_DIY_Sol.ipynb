**Q1. Problem Statement: Probability Distribution**

Load the exam.csv dataset into a DataFrame and perform the following tasks on the “math score,” ”readingscore,” and “writing score” columns:

1.Plot the distributions using the seabornlibrary

2.Find the mean and standard deviation using the scipylibrary

3.Find skewness and kurtosisusing thestatsmodule from the scipylibrary

4.Find Standard Normal Distribution(z-score) using the statsmodule from the scipylibrary

5.Plot the Z distribution using the seabornlibrary

```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("exam.csv")
df
```

```python
#1.Plot the distributions
sns.histplot(df["math score"], kde=True)
plt.title("Math Score Distribution")
plt.show()

sns.histplot(df["reading score"], kde= True)
plt.title("Reading score Distribution")
plt.show()

sns.histplot(df["writing score"], kde=True)
plt.title("Writing Score Distribution")
plt.show()
```

```python
#2. find mean and standard deviation of Math score
mean_math = np.mean(df["math score"])
std_math = np.std(df["math score"])
print("Mean of Math Score:", mean_math,)
print("Standard Deviation of math Score:", std_math, "\n")

#find mean and standard deviation of reading score
mean_reading = np.mean(df["reading score"])
std_reading = np.std(df["reading score"])
print("Mean of Reading Score:", mean_reading)
print("Standard Deviation Score:", std_reading,"\n")

#find mean and standard deviation of writing swcore
mean_writing = np.mean(df["writing score"])
std_writing = np.std(df["writing score"])
print("Mean of Writing Score:", mean_writing)
print("Standard Deviation Score:", std_writing)
```

```python
#3. Find Skewness and kurtosis of math score
skew_math = stats.skew(df["math score"])
kurt_math = stats.kurtosis(df["math score"])
print("Skewness of Math Score:", skew_math)
print("Kurthosis of Math Score:", kurt_math,"\n")

#find skewness and kurtosis of reading score
skew_reading= stats.skew(df["reading score"])
kurt_reading = stats.kurtosis(df["reading score"])
print("Skewness of Reading Score:", skew_reading)
print("Kurthosis of Reading Score:", kurt_reading,"\n")

#find skewness and kurtosis of writing score
skew_writing = stats.skew(df["writing score"])
kurt_writing = stats.kurtosis(df["writing score"])
print("Skewness of Writing Score:", skew_writing)
print("Kurthosis of Writing Score:", kurt_writing)
```

```python
#4. Find Z-Scores of math, reading, writing scores
z_scores_math = (df["math score"] - mean_math)/ std_math
z_scores_reading =(df["reading score"] - mean_reading)/ std_reading
z_scores_writing = (df["writing score"] - mean_writing)/std_writing
print("Z-scores of Math Score:", z_scores_math)
print("Z-scores of Reading Score:",z_scores_reading)
print("Z-scores of Writing Score:", z_scores_writing)
```

```python
#5. Plot Z-distribution of math,reading,writing scores
sns.histplot(z_scores_math, kde= True)
plt.title("Z-Distribution for Math Scores")
plt.show()
sns.histplot(z_scores_reading, kde= True)
plt.title("Z-Distribution for Reading Scores")
plt.show()
sns.histplot(z_scores_writing, kde= True)
plt.title("Z-Distribution for Writing Scores")
plt.show()
```
